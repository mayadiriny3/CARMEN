## Representación con armonicos 

### Calcular la serie de Fourier con 10 armónicos
### N = 10

```python
# Definir la función f(x)
def f(x):
    if -np.pi <= x <= 0:
        return 0
    elif 0 <= x <= np.pi:
        return x

# Calcular los coeficientes de Fourier
def fourier_coefficients(n):
    a0 = 1 / (2 * np.pi)
    if n == 0:
        an = 0
    else:
        an = (2 / (n * np.pi)) * (1 - np.cos(n * np.pi))
    bn = 0
    return a0, an, bn

# Calcular la serie de Fourier truncada
def fourier_series(x, N):
    series_sum = 0
    a0, an, bn = fourier_coefficients(0)
    series_sum += a0 / 2
    for n in range(1, N+1):
        an, bn = fourier_coefficients(n)[1:]
        series_sum += an * np.cos(n * x) + bn * np.sin(n * x)
    return series_sum

# Generar los valores de x y f(x) para graficar
x = np.linspace(-np.pi, np.pi, 1000)
y = np.array([f(xi) for xi in x])

# Calcular y graficar la serie de Fourier truncada
N = 10  # Número de términos de la serie truncada
y_fourier = np.array([fourier_series(xi, N) for xi in x])

plt.plot(x, y, label='f(x)')
plt.plot(x, y_fourier, label=f'Fourier (N={N})')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Serie de Fourier')
plt.legend()
plt.grid(True)
plt.show()
```

## Representación gráfica 

```python
##10
import numpy as np
import matplotlib.pyplot as plt

# Definir la función original
def f(x):
    return x**2

# Definir el número máximo de armónicos
n_max = 10

# Definir el intervalo de valores de x
x = np.linspace(-np.pi, np.pi, 1000)

# Calcular los coeficientes a_n
a_n = np.zeros(n_max)
for n in range(1, n_max + 1):
    integrand = lambda x: f(x) * np.cos(n * x)
    a_n[n-1] = 1/np.pi * np.trapz(integrand(x), x)

# Calcular la serie truncada de Fourier
f_n = np.zeros_like(x)
for n in range(1, n_max + 1):
    f_n += a_n[n-1] * np.cos(n * x)

# Graficar la función original y la serie truncada
plt.plot(x, f(x), label='Función original')
plt.plot(x, f_n, label=f'Serie truncada ({n_max} armónicos)')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.title('Convergencia de la serie de Fourier')
plt.grid(True)
plt.show()
```
